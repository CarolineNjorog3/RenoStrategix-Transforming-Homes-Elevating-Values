##Data cleaning

```python
#Checking for null values
pd.DataFrame(data.isna().sum()).T
```

```python
# Replacing the Null values at Waterfront with "No"
data['waterfront'].fillna('NO', inplace=True)
```

```python
# Replacing the null values in the year renovated with the
# values in the year built
data['yr_renovated'].fillna(data['yr_built'], inplace=True)
```

The assumption made in this case was that the datapoints where there was a NaN value represented houses that had never been renovated.

```python
# Dropping the null values in the view column
data.dropna(inplace=True)
```

```python
# Replacing "?" with 0 in sqft_basement column
data['sqft_basement'] = data['sqft_basement'].replace('?', 'None', inplace=True)
```

The assumption made in this case was that the datapoints where there was a '?' represented houses that have no basement.

```python
# Checking for null values after cleaning the data
pd.DataFrame(data.isna().sum()).T
```

```python
data.info
```

```python
# Creating a new column to find the difference between the year renovated and year built
# 0 Has been replaced for the rows that returned -ve years.
data['Years_Since_Renovation'] = data.apply(lambda row: row['yr_renovated'] - row['yr_built'] if row['yr_renovated'] > 0 else 0, axis=1)
```

```python
# Checking for any negative values in the new column.
if (data['Years_Since_Renovation'] < 0).any():
    print("There are negative values in the 'Years_Since_Renovation' column.")
else:
    print("There are no negative values in the 'Years_Since_Renovation' column.")
```

```python
#Checking the data types in the data frame
data.info()
```

```python
#Changing the years data type to int
data['yr_built'] = data['yr_built'].astype(int)
data['yr_renovated'] = data['yr_renovated'].astype(int)
data['Years_Since_Renovation'] = data['Years_Since_Renovation'].astype(int)

#Changing the date column data type to datetime.
data['date'] = pd.to_datetime(data['date'])
```

#  Simple Linear Regression Modelling

```python
# Import the necessary libraries for statistical modeling
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming 'data' is your DataFrame
# Define the dependent variable 'y' and the independent variable 'x'
y = data['price']
x = data['sqft_living']

# Add a constant term to the independent variable
x = sm.add_constant(x)

# Create a linear regression model using Ordinary Least Squares (OLS)
model = sm.OLS(y, x).fit()

# Print the summary of the regression
print(model.summary())
```

### OLS Regression Summary Explanation

#### Model Fit:
- **R-squared:** 0.492 (49.2% explained variability)
  - The R-squared value indicates that nearly half of the variability in house prices is explained by the model, suggesting a moderate level of predictive power.
- **F-statistic:** 2.087e+04 (p-value: 0.00)
  - The low p-value associated with the F-statistic implies that the overall model is statistically significant, providing evidence that at least one predictor is related to the response variable.

#### Coefficients:
- **Intercept (`const`):** -4.215e+04
  - The intercept represents the estimated house price when the square footage is zero, though in this context, this might not be practically meaningful.
- **sqft_living:** 279.9321
  - The coefficient for 'sqft_living' indicates that, on average, for every one-unit increase in square footage, the predicted house price increases by $279.9321.

#### Hypothesis Testing:
- Coefficient p-values: Very close to zero
  - The very low p-values associated with both the intercept and 'sqft_living' indicate that both coefficients are statistically significant. This means there is strong evidence to reject the null hypothesis that the corresponding coefficients are zero.


#### Conclusion:
In summary, the regression model suggests that the square footage of living space (`sqft_living`) significantly influences house prices (`price`). The model demonstrates a moderate level of explanatory power, and some diagnostics indicate potential issues that warrant further investigation.
